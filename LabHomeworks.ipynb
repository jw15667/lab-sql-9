```python
engine.execute("USE sakila")
query = 'SELECT c.customer_id, c.first_name, c.last_name, c.email, r.rental_id, count(c.customer_id) as total_rentals \
FROM rentals_may as r \
JOIN customer as c \
ON c.customer_id = r.customer_id \
GROUP BY c.customer_id \
ORDER BY total_rentals desc'
data_may = pd.read_sql_query(query, engine)
data_may.head()
```

Check the number of rentals for each customer for June

```python
engine.execute("USE sakila")
query = 'SELECT c.customer_id, c.first_name, c.last_name, c.email, r.rental_id, count(c.customer_id) as total_rentals \
FROM rentals_june as r \
JOIN customer as c \
ON c.customer_id = r.customer_id \
GROUP BY c.customer_id \
ORDER BY total_rentals desc'
data_june = pd.read_sql_query(query, engine)
data_june.head()
```

Write a function that checks if customer borrowed more or less books in the month of June as compared to May.

```python
engine.execute("USE sakila")
query = 'SELECT m.customer_id, j.rental_id, count(j.customer_id) as june_total_rentals, count(m.customer_id) as may_total_rentals \
FROM rentals_june as j \
JOIN rentals_may as m \
ON j.customer_id=m.customer_id \
GROUP BY j.customer_id ;'
data_may_june = pd.read_sql_query(query, engine)
data_may_june.head()
```

```python
def check_more_or_less_dvds(customer_id, dataframe1, dataframe2):
    June = dataframe1.loc[dataframe1['customer_id']==customer_id]
    May = dataframe2.loc[dataframe2['customer_id']==customer_id]
    return pd.DataFrame([[int(June.iloc[0, 5]) - int(May.iloc[0, 5])]], columns=['June Compared to May, Customer {}'.format(customer_id)])
```
